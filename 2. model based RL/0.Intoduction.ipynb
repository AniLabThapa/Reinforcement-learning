# Policy Iteration Algorithm

Policy iteration is an iterative method used to find the optimal policy in a Markov Decision Process (MDP).

It consists of two main steps:

## 1. Policy Evaluation
Compute the value function for the current policy $\pi$:

$$
V(s) \leftarrow \sum_{s', r} P(s', r \mid s, \pi(s)) \left[r + \gamma V_{\text{old}}(s')\right]
$$

for all $s \in S$, repeated until convergence.

## 2. Policy Improvement
Update the policy by choosing the action that gives the best expected return:

$$
\pi(s) \leftarrow \arg\max_a \sum_{s', r} P(s', r \mid s, a) \left[r + \gamma V(s')\right]
$$

for all $s \in S$.

Repeat the two steps above until the policy no longer changes.

---

# Value Iteration Algorithm

Value iteration combines policy evaluation and policy improvement into one step.

Repeat the following update for all $s \in S$:

$$
V(s) \leftarrow \max_a \sum_{s', r} P(s', r \mid s, a) \left[r + \gamma V_{\text{old}}(s')\right]
$$

Continue until convergence.

At the end, the optimal value function is obtained:

$$
V_*(s) = V(s) \quad \text{for all } s \in S
$$

Then the optimal policy is extracted as:

$$
\pi(s) \leftarrow \arg\max_a \sum_{s', r} P(s', r \mid s, a) \left[r + \gamma V_*(s')\right]
$$

---

# Key Difference

- **Policy Iteration**: alternates between policy evaluation and policy improvement
- **Value Iteration**: directly updates the value function using the Bellman optimality backup

---

# Notation

- $S$: set of states
- $s$: a state
- $a$: an action
- $r$: reward
- $P(s', r \mid s, a)$: transition probability
- $\gamma$: discount factor
- $V(s)$: state-value function
- $\pi(s)$: policy
- $V_*(s)$: optimal value function


# Limitation —

## 1. Synchronous updates
- operation over entire state-space of MDP
- **Bad! very bad computationally.**  
  *(e.g. Chess)*

## 2. Need knowledge of the environment dynamics
- need to know the transition model:
  
  $$
  P(s', r \mid s, a)
  $$

## 3. No "learn from experience / interaction"
- cannot learn directly from interaction with the environment
- requires a known model of the MDP instead of experience-based learning
